# Bike

```python
import pandas as pd
import numpy as np
save_dir = "D:\Study\自学\表格数据生成/v11\Dataset/bike"
df = pd.read_csv(f'{save_dir}/hour.csv')

def replace_season(value):
    if value == 1:
        return 'winter'
    elif value == 2:
        return 'spring'
    elif value == 3:
        return 'summer'
    elif value == 4:
        return 'fall'
    else:
        return value

def replace_year(value):
    if value == 0:
        return 2011
    else:
        return 2012
    
def replace_month(value):
    if value == 1:
        return 'January'
    elif value == 2:
        return 'February'
    elif value == 3:
        return 'March'
    elif value == 4:
        return 'April'
    elif value == 5:
        return 'May'
    elif value == 6:
        return 'June'
    elif value == 7:
        return 'July'
    elif value == 8:
        return 'August'
    elif value == 9:
        return 'September'
    elif value == 10:
        return 'October'
    elif value == 11:
        return 'November'
    elif value == 12:
        return 'December'
    
def replace_weekday(value):
    if value == 0:
        return 'Sunday'
    elif value == 1:
        return 'Monday'
    elif value == 2:
        return 'Tuesday'
    elif value == 3:
        return 'Wednesday'
    elif value == 4:
        return 'Thursday'
    elif value == 5:
         return 'Friday'
    elif value == 6:
         return 'Saturday'

def replace_weathersit(value):
    if value == 1:
        return 'Clear'
    elif value == 2:
        return 'Mist'
    elif value == 3:
        return 'Light Snow'
    elif value == 4:
        return 'Heavy Rain'
```

```python
df['season'] = df['season'].apply(replace_season)
df['yr'] = df['yr'].apply(replace_year)
df['mnth'] = df['mnth'].apply(replace_month)
df['weekday'] = df['weekday'].apply(replace_weekday)
df['weathersit'] = df['weathersit'].apply(replace_weathersit)
df.head()
```

```python
df.to_csv(f"{save_dir}/bike.csv", index=False)
```

```python
save_dir = "D:\Study\自学\表格数据生成/v11\Dataset/productivity/real"
np.save(f'{save_dir}/X_num_train.npy', X_num_train)
np.save(f'{save_dir}/X_cat_train.npy', X_cat_train)
np.save(f'{save_dir}/Y_train.npy', y_train)

np.save(f'{save_dir}/X_num_test.npy', X_num_test)
np.save(f'{save_dir}/X_cat_test.npy', X_cat_test)
np.save(f'{save_dir}/Y_test.npy', y_test)

np.save(f'{save_dir}/X_num_val.npy', X_num_val)
np.save(f'{save_dir}/X_cat_val.npy', X_cat_val)
np.save(f'{save_dir}/Y_val.npy', y_val)

train_df.to_csv(f"{save_dir}/train.csv", index=False)
test_df.to_csv(f"{save_dir}/test.csv", index=False)
val_df.to_csv(f"{save_dir}/val.csv", index=False)
```
