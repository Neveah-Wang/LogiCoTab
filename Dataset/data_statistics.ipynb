# Adult

```python
Y = np.load("D:\Study\自学\表格数据生成/v11\Dataset/adult\Y_train.npy", allow_pickle=True)
X_num = np.load("D:\Study\自学\表格数据生成/v11\Dataset/adult\X_num_train.npy", allow_pickle=True).astype(np.float32)
X_cat = np.load("D:\Study\自学\表格数据生成/v11\Dataset/adult\X_cat_train.npy", allow_pickle=True).astype(str)
```

```python
print(Y.shape)
print(X_num.shape)
print(X_cat.shape)
```

```python
X_num_columns = [
    "age",
    "fnlwgt",
    "education_num",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
]
X_cat_columns = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]
y_column = [
    "salary",
]

x_df = pd.DataFrame(np.concatenate((X_num, X_cat), axis=1), columns=X_num_columns + X_cat_columns)
y_df = pd.DataFrame(Y, columns=y_column)
merged_df = pd.concat([x_df, y_df], axis=1)
```
