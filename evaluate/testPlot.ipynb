## 计算相关矩阵(correlation matrices)并绘制 heatmap：
1. corr()
    
    教程：https://blog.csdn.net/qq_38614074/article/details/139654894

    主要使用的到了pandas中的函数 corr()，该函数可以计算Dataframe中不同列之间的相关系数。

    皮尔逊相关系数（Pearson correlation coefficient），即概率论课上学的相关系数，可以用于判断两个变量之间是否存在线性相关。取值范围为[-1,1]，等于1时，说明两个变量成线性相关，等于0时说明两个变量没有线性关系。

2. seaborn 
    
    教程：https://blog.csdn.net/qq_43201025/article/details/131042439

    用于绘制可视化的图

```python
# Calculate correlation matrices and their absolute differences
real_corr = original_data.corr()
generated_corr = generated_data.corr()
diff_corr = np.abs(real_corr - generated_corr)


# Replace NaN values with 0
diff_corr = diff_corr.fillna(0)

# Plot heatmap of the absolute difference
plt.figure(figsize=(10, 8))
# sns.heatmap(diff_corr, cmap='Reds', cbar=False, vmin=0, vmax=2, annot=True, fmt=".2f")
sns.heatmap(diff_corr, cmap='Reds', cbar=True, vmin=0, vmax=1.5, annot=False)
plt.title('Absolute difference between correlation matrices', fontsize=20)
plt.xlabel('Features', fontsize=16)
plt.ylabel('Features', fontsize=16)
# 设置刻度值的大小
plt.tick_params(axis='both', which='major', labelsize=16)

# Save plot
heatmap_filename = os.path.join(save_folder, 'correlation_heatmap.png')
plt.tight_layout()
plt.savefig(heatmap_filename)
plt.close()

# Calculate and print the sum of all values in the difference correlation matrix
total_sum = diff_corr.values.sum()
print(f'Sum of all values in the correlation heatmap: {total_sum}')
print(f'max value: {np.max(diff_corr)}')
print(f'min value: {np.min(diff_corr)}')

# Save the sum to a file
sum_filename = os.path.join(save_folder, 'metrics.txt')
with open(sum_filename, 'a') as f:
    f.write(f'Sum of all values in the correlation heatmap: {total_sum}'+'\n')
```

#### 对于 numerical 数据，计算 Wasserstein distance

```python
original_data = pd.DataFrame(X_num_train, columns=X_num_columns)
generated_data = pd.DataFrame(X_num_synthesis, columns=X_num_columns)


# Compute Wasserstein distance using original data
wasserstein_dist = wasserstein_distance(original_data.values.flatten(), generated_data.values.flatten())
# wasserstein_dist = wasserstein_distance(original_data.values, generated_data.values)
print(wasserstein_dist)
```

观察下面的结果可以发现：

* 如果改变数据的分布顺序，是不会影响结果的；
* 如果增加数据，会影响分布情况

因此，在测试时候，应该成成跟原数据相同分数的情况，不能对数据进行平衡操作。

```python
# test
print(wasserstein_distance([1,2,0, 0,1,1, 2,1,0], [1,2,0, 0,1,1, 2,1,0]))
print(wasserstein_distance([1,2,0, 0,1,1, 2,1,0], [1,2,0, 2,1,0, 0,1,1]))
print(wasserstein_distance([1,2,0, 0,1,1, 2,1,0], [1,2,0, 2,1,0, 0,1,1, 1,2,0]))
print(wasserstein_distance([1,2,0, 0,1,1, 2,1,0], [0,1,1, 1,2,0, 2,1,0, 1,2,0]))
```

```python
original_data = pd.DataFrame(X_cat_train_encoded, columns=X_cat_columns)
generated_data = pd.DataFrame(X_cat_synthesis_encoded, columns=X_cat_columns)

# 先观察一下绘制成直方图后的数据分布情况
plt.hist(original_data.values.flatten(), bins=30, edgecolor='black')
plt.show()
plt.hist(generated_data.values.flatten(), bins=30, edgecolor='black')
plt.show()

# Compute JS distance using original data
# Compute histograms with the same bins
hist_original, _ = np.histogram(original_data.values.flatten(), bins=30, density=True)   # density=True 则返回频率密度而不是计数，即直方图面积之和为1
hist_generated, _ = np.histogram(generated_data.values.flatten(), bins=30, density=True)


# Compute JS distance
js_dist = np.sqrt(0.5 * (entropy(hist_original, (hist_original + hist_generated) / 2) + 
                         entropy(hist_generated, (hist_original + hist_generated) / 2)))
print(js_dist)
```
